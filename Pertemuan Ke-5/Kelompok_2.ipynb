```python
from dataclasses import dataclass
import time

# ==========================================
# 1. PARENT CLASSES (Ciri Umum)
# ==========================================

@dataclass
class PenyakitPernapasan:
    """Parent class untuk penyakit yang utamanya ditandai dengan Batuk/Sesak"""
    nama: str = "Penyakit Saluran Pernapasan"
    gejala_utama: str = "Batuk / Sesak Napas"
    sistem_diserang: str = "Saluran Pernapasan (Paru-paru/Tenggorokan)"
    penularan: str = "Udara / Droplet"

@dataclass
class InfeksiSistemik:
    """Parent class untuk penyakit yang utamanya ditandai dengan Demam"""
    nama: str = "Infeksi Sistemik"
    gejala_utama: str = "Demam"
    sistem_diserang: str = "Aliran Darah / Pencernaan"
    penularan: str = "Bakteri / Vektor Nyamuk"

# ==========================================
# 2. CHILD CLASSES (Penyakit Spesifik / Knowledge Base)
# ==========================================

# -- Kategori Pernapasan --
@dataclass
class Covid19(PenyakitPernapasan):
    nama: str = "COVID-19"
    ciri_khas: str = "Kehilangan indra penciuman (Anosmia)"
    bahaya: str = "Tinggi (Sangat Menular)"

@dataclass
class TBC(PenyakitPernapasan):
    nama: str = "Tuberkulosis (TBC)"
    ciri_khas: str = "Batuk berdarah dan berlangsung > 2 minggu"
    bahaya: str = "Tinggi (Butuh pengobatan 6 bulan)"

# -- Kategori Sistemik --
@dataclass
class DBD(InfeksiSistemik):
    nama: str = "Demam Berdarah Dengue (DBD)"
    ciri_khas: str = "Muncul bintik merah pada kulit dan nyeri sendi"
    bahaya: str = "Gawat Darurat (Risiko Syok)"

@dataclass
class Tifus(InfeksiSistemik):
    nama: str = "Demam Tifoid (Tifus)"
    penularan: str = "Makanan/Air terkontaminasi (Fecal-oral)" # Override atribut Parent
    ciri_khas: str = "Demam naik secara perlahan, memburuk di malam hari"
    bahaya: str = "Sedang"

# ==========================================
# 3. MESIN INFERENSI (Logika Diagnosa)
# ==========================================

def jalankan_sistem_pakar():
    print("\n" + "="*50)
    print(" 🏥 SISTEM PAKAR DIAGNOSA MEDIS (KELOMPOK 2) 🏥 ")
    print("="*50)
    print("Pilih Gejala Utama yang dialami:")
    print("1. Batuk / Sesak Napas")
    print("2. Demam Tinggi")

    pilihan = input("Masukkan angka (1/2): ").strip()

    print("\nSedang menganalisis ciri khusus...")
    time.sleep(1) # Efek loading agar demo lebih aesthetic

    if pilihan == "1":
        print("\nPilih ciri khusus yang paling sesuai:")
        print("1. Kehilangan indra penciuman (Anosmia)")
        print("2. Batuk berdarah / berlangsung > 2 Minggu")
        ciri = input("Masukkan angka (1/2): ").strip()

        # Logika menggunakan angka
        hasil = Covid19() if ciri == "1" else TBC()

    elif pilihan == "2":
        print("\nPilih ciri khusus yang paling sesuai:")
        print("1. Muncul bintik merah pada kulit")
        print("2. Demam memburuk saat malam hari")
        ciri = input("Masukkan angka (1/2): ").strip()

        # Logika menggunakan angka
        hasil = DBD() if ciri == "1" else Tifus()

    else:
        print("Gejala tidak valid. Harap masukkan angka 1 atau 2.")
        return

    # ==========================================
    # 4. OUTPUT HASIL DIAGNOSA
    # ==========================================
    print("\n" + "-"*50)
    print(f"📌 DIAGNOSIS     : {hasil.nama.upper()}")
    print("-"*50)
    print(f"📖 Kategori      : {hasil.__class__.__bases__[0].__name__}")
    print(f"🦠 Penularan     : {hasil.penularan}")
    print(f"🎯 Fokus Infeksi : {hasil.sistem_diserang}")
    print(f"🔍 Ciri Khas     : {hasil.ciri_khas}")
    print(f"⚠️ Bahaya        : {hasil.bahaya}")
    print("="*50 + "\n")

# Menjalankan program
if __name__ == "__main__":
    jalankan_sistem_pakar()
```

```python
from dataclasses import dataclass
import time

# ==========================================
# 1. PARENT CLASS (Gejala Utama Demam)
# ==========================================
@dataclass
class PenyakitDemam:
    """Parent class: Semua penyakit di bawah ini wajib punya gejala Demam"""
    nama: str = "Penyakit Umum"
    gejala_utama: str = "Demam"

# ==========================================
# 2. CHILD CLASSES (Database Penyakit)
# ==========================================
@dataclass
class Flu(PenyakitDemam):
    nama: str = "Influenza (Flu)"
    ciri_khas: str = "Hidung meler/pilek, bersin-bersin, dan nyeri otot/sendi"
    bahaya: str = "Rendah (Butuh istirahat dan vitamin)"

@dataclass
class ISPA(PenyakitDemam):
    nama: str = "Infeksi Saluran Pernapasan Akut (ISPA)"
    ciri_khas: str = "Sakit tenggorokan, batuk, dan napas terasa berat"
    bahaya: str = "Sedang (Bisa memicu infeksi paru-paru)"

@dataclass
class Tifus(PenyakitDemam):
    nama: str = "Demam Tifoid (Tifus)"
    ciri_khas: str = "Suhu badan naik perlahan, memburuk saat malam hari disertai mual"
    bahaya: str = "Sedang hingga Tinggi (Butuh antibiotik)"

@dataclass
class DBD(PenyakitDemam):
    nama: str = "Demam Berdarah Dengue (DBD)"
    ciri_khas: str = "Demam tinggi mendadak disertai munculnya bintik merah di kulit"
    bahaya: str = "Sangat Tinggi (Gawat Darurat / Risiko Syok)"

# ==========================================
# 3. MESIN INFERENSI (Logika Diagnosa)
# ==========================================
def jalankan_sistem_pakar():
    print("\n" + "="*65)
    print(" 🏥 SISTEM PAKAR DIAGNOSA MEDIS (FOKUS GEJALA DEMAM) 🏥 ")
    print("="*65)
    print("Apakah pasien mengalami gejala utama DEMAM?")
    print("1. Ya")
    print("2. Tidak")

    utama = input("Masukkan angka (1/2): ").strip()
    print("-" * 65)

    if utama == "1":
        print("Pilih gejala penyerta (ciri khas) yang dialami pasien:")
        print("1. Pilek, bersin-bersin, dan badan pegal      ")
        print("2. Sakit tenggorokan, batuk, dan napas berat  ")
        print("3. Demam memburuk saat malam hari & mual      ")
        print("4. Demam tinggi mendadak & bintik merah       ")

        ciri = input("Masukkan angka (1-4): ").strip()

        if ciri == "1": hasil = Flu()
        elif ciri == "2": hasil = ISPA()
        elif ciri == "3": hasil = Tifus()
        elif ciri == "4": hasil = DBD()
        else:
            print("Input tidak valid. Harap masukkan angka 1-4.")
            return

    elif utama == "2":
        print("Sistem pakar Kelompok 2 ini dikhususkan untuk mendiagnosis penyakit dengan basis Demam.")
        return

    else:
        print("Input tidak valid. Harap masukkan angka 1 atau 2.")
        return

    # Efek delay agar terlihat sistem sedang memproses klasifikasi
    print("\nSedang menganalisis basis pengetahuan...")
    time.sleep(1)

    # ==========================================
    # 4. OUTPUT HASIL DIAGNOSA
    # ==========================================
    print("\n" + "="*65)
    print(f"📌 HASIL DIAGNOSIS : {hasil.nama.upper()}")
    print("="*65)
    print(f"🤒 Gejala Dasar  : {hasil.gejala_utama}")
    print(f"🔍 Ciri Khas     : {hasil.ciri_khas}")
    print(f"⚠️ Bahaya        : {hasil.bahaya}")
    print("="*65 + "\n")

# Menjalankan program
if __name__ == "__main__":
    jalankan_sistem_pakar()
```
